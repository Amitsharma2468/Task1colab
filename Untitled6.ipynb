```python
!pip install pandas
import matplotlib.pyplot as plt
import pandas as pd  # Import pandas

def visualize_temperature_trend(data):
    dates = data['Date']
    temperatures = data['Temperature (°C)']

    plt.figure(figsize=(10, 6))
    plt.plot(dates, temperatures, marker='o', linestyle='-', color='b')
    plt.xlabel("Date")
    plt.ylabel("Temperature (°C)")
    plt.title("Temperature Trend")
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def visualize_daily_rainfall(data):
    dates = data['Date']
    rainfall = data['Rainfall (mm)']

    plt.figure(figsize=(10, 6))
    plt.bar(dates, rainfall, color='skyblue')
    plt.xlabel("Date")
    plt.ylabel("Rainfall (mm)")
    plt.title("Daily Rainfall")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


# Load weather data using pandas
weather_data = pd.read_csv('8_Weather_Data_Analysis_Tool.csv')

visualize_temperature_trend(weather_data)
visualize_daily_rainfall(weather_data)
```

```python
import csv

def load_weather_data(filename):

  data = []
  with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
      data.append(row)
  return data

def validate_date(date_str):

  try:
    year, month, day = map(int, date_str.split('-'))
    if 1 <= month <= 12 and 1 <= day <= 31:
      return True
    else:
      return False
  except ValueError:
    return False

def validate_numerical_field(value):

  try:
    float(value)
    return True
  except ValueError:
    return False

def calculate_average(data, field):

  if len(data) == 0:
    return None
  total = 0
  for row in data:
    if field in row and validate_numerical_field(row[field]):
      total += float(row[field])
  return total / len(data)

def find_max_value(data, field):

  max_value = None
  for row in data:
    if field in row and validate_numerical_field(row[field]):
      value = float(row[field])
      if max_value is None or value > max_value:
        max_value = value
  return max_value

def find_min_value(data, field):

  min_value = None
  for row in data:
    if field in row and validate_numerical_field(row[field]):
      value = float(row[field])
      if min_value is None or value < min_value:
        min_value = value
  return min_value

def count_days_by_condition(data, condition):

  count = 0
  for row in data:
    if 'Condition' in row and row['Condition'] == condition:
      count += 1
  return count

def generate_summary_report(data, start_date, end_date):

  filtered_data = [row for row in data if start_date <= row['Date'] <= end_date]

  report = {
      'Average Temperature': calculate_average(filtered_data, 'Temperature (°C)'),
      'Max Temperature': find_max_value(filtered_data, 'Temperature (°C)'),
      'Min Temperature': find_min_value(filtered_data, 'Temperature (°C)'),
      'Average Rainfall': calculate_average(filtered_data, 'Rainfall (mm)'),
      'Max Rainfall': find_max_value(filtered_data, 'Rainfall (mm)'),
      'Average Humidity': calculate_average(filtered_data, 'Humidity (%)'),
      'Average Wind Speed': calculate_average(filtered_data, 'Wind Speed (km/h)')
  }
  return report

# Example usage:
filename = '8_Weather_Data_Analysis_Tool.csv'
weather_data = load_weather_data(filename)

average_temp = calculate_average(weather_data, 'Temperature (°C)')
print(f"Average Temperature: {average_temp:.2f} °C")

max_rainfall = find_max_value(weather_data, 'Rainfall (mm)')
print(f"Maximum Rainfall: {max_rainfall} mm")

sunny_days = count_days_by_condition(weather_data, 'Sunny')
print(f"Number of Sunny Days: {sunny_days}")

start_date = '2025-01-01'
end_date = '2025-01-31'
report = generate_summary_report(weather_data, start_date, end_date)
for key, value in report.items():
  print(f"{key}: {value:.2f}" if isinstance(value, float) else f"{key}: {value}")
```

```python

import csv
import matplotlib.pyplot as plt
import json
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display, clear_output

def export_to_csv(data, filename):
    try:
        with open(filename, 'w', newline='') as csvfile:
            fieldnames = data[0].keys() if data else []
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(data)
        print(f"Data exported to {filename} successfully.")
    except Exception as e:
        print(f"Error exporting data: {e}")

def save_config(config, filename):
    try:
        with open(filename, 'w') as f:
            json.dump(config, f)
        print(f"Configuration saved to {filename}")
    except Exception as e:
        print(f"Error saving configuration: {e}")

def load_config(filename):
    try:
        with open(filename, 'r') as f:
            config = json.load(f)
            return config
    except FileNotFoundError:
        print(f"Configuration file not found: {filename}")
        return {}
    except json.JSONDecodeError:
        print(f"Invalid JSON format in configuration file: {filename}")
        return {}

def interactive_search(data):
    def search_data(date_input, condition_input):
        clear_output(wait=True)
        results = []

        for row in data:
            if date_input and row['Date'] == date_input:
                results.append(row)
            elif condition_input and 'Condition' in row and row['Condition'] == condition_input:
                results.append(row)

        if not results:
            print("No matching data found.")
        else:
            for row in results:
                print(row)

    date_widget = widgets.Text(description='Date (YYYY-MM-DD):', placeholder='Enter Date')
    condition_widget = widgets.Dropdown(description='Condition:', options=['', 'Sunny', 'Cloudy', 'Rainy', 'Windy'] + list(set([row['Condition'] for row in data if 'Condition' in row])))
    interact(search_data, date_input=date_widget, condition_input=condition_widget);


interactive_search(weather_data)
```
